## **SPLIT THE DATA**

```python
X = df.drop(['benign_0__mal_1'],axis=1).values
```

```python
y = df['benign_0__mal_1'].values
```

```python
y
```

## **TRAINING THE DATA**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 101, test_size = 0.25)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

## **SCALE the DATA**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
Scaler = MinMaxScaler()
```

```python
X_train = Scaler.fit_transform(X_train)
```

```python
X_test = Scaler.transform(X_test)
```

## **BUILDING MODEL**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model = Sequential()

# Hidden Layer
model.add(Dense(30, activation = 'relu'))
model.add(Dense(15, activation = 'relu'))

# Output Layer
model.add(Dense(1, activation = 'softmax'))

model.compile(optimizer='adam', loss='binary_crossentropy')
```

## **FIT our Data**

```python
model.fit(x = X_train, y = y_train, epochs=600, validation_data =(X_test, y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

**Conclusion :** Orange line indicates Data is Overfitting


**Solution :** We need to restrict this by Early Stopping Callbacks

## **CALLBACKS**

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# help(EarlyStopping)
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=25, verbose=1, mode='min')
```

## **REDEFINE MODEL - EARLY STOPPING to Prevent losses**

```python
model = Sequential()

# Hidden Layer
model.add(Dense(30, activation = 'relu'))
model.add(Dense(15, activation = 'relu'))

# Output Layer
model.add(Dense(1, activation = 'softmax'))

model.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
model.fit(x = X_train, y = y_train, epochs=600, validation_data =(X_test, y_test), callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

**Conclusion** : EarlyStop helps to prevent Overfitting

## **REDEFINE MODEL - DROPOUT to Prevent losses**

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()

# Hidden Layer
model.add(Dense(30, activation = 'relu'))
model.add(Dropout(0.5))

model.add(Dense(15, activation = 'relu'))
model.add(Dropout(0.5))

# Output Layer
model.add(Dense(1, activation = 'softmax'))

model.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=25, verbose=1, mode='min')
```

```python
model.fit(x = X_train, y = y_train, epochs=600, validation_data =(X_test, y_test), callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
model_loss.plot()
```

**Conclusion :** Dropout affects and works better to train our model

## **PREDICTION**

```python
y_pred = model.predict(X_test)

y_pred = (y_pred>=0.5)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix

confusion_matrix(y_test, y_pred)
```
